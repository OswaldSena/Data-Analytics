```python
#Using DataReader as a webscraping to tool to get stock info the above banks between 2006 and 2016 from Yahoo Finance
```

```python
# Bank of America
BAC = web.DataReader('BAC', 'yahoo', start='2006-01-01', end='2016-12-31')

# CitiGroup
C = web.DataReader('C', 'yahoo', start='2006-01-01', end='2016-12-31')

# Goldman Sachs
GS = web.DataReader('GS', 'yahoo', start='2006-01-01', end='2016-12-31')

# JPMorgan Chase
JPM = web.DataReader('JPM', 'yahoo', start='2006-01-01', end='2019-12-31')
# Morgan Stanley
MS = web.DataReader('MS', 'yahoo', start='2006-01-01', end='2016-12-31')

# Wells Fargo
WFC = web.DataReader('WFC', 'yahoo', start='2006-01-01', end='2016-12-31')

#Each bank was setup to be a different DataFrame in this project with their ticker symbols being used to define the DataFrame
```

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

```python
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC],axis=1,keys=tickers)
```

Setting the column name levels (this is filled out for you)

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

# Exploratory Data Analysis


** Max Close price for each bank's stock throughout the time period**

```python
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

** Creating a function to get the returns of each Bank over the period:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()
```

```python
for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change()
returns.head()

#This loop iterates through each Bank Stock Ticker and creates a return column
```

** Pairplot of the returns DataFrame using seaborn**

```python
#returns[1:]
import seaborn as sns
sns.pairplot(returns[1:])
```

## Moving Averages

Let's analyze the moving averages for these stocks in the year 2008. 

** Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
plt.figure(figsize=(12,6))
BAC['Close'].loc['2008-01-01':'2009-01-01'].rolling(window=30).mean().plot(label='30 Day Avg')
BAC['Close'].loc['2008-01-01':'2009-01-01'].plot(label='BAC CLOSE')
plt.legend()
```

** Heatmap of the correlation between the stocks Close Price.**


```python
sns.heatmap(bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr(),annot=True)
```

** Using seaborn's clustermap to cluster the correlations together:**



```python
sns.clustermap(bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr(),annot=True)
```
